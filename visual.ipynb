```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('cleaned.csv')
```

```python
colors =['#EDAF97', '#FE938C', '#C49792', '#AD91A3', '#9D91A3']
genres = df['status'].value_counts().nlargest(20).plot(kind='pie', colors=colors)
genres.set_title('Статус')
```
